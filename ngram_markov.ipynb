```python
import numpy as np
```

```python
import pandas as pd

data = pd.read_csv('output.txt', sep='\n\n\n', engine='python',encoding='utf8', header = 0)
```

```python
data.head()
```

```python
linelist= data["Artist: 40 Thevz f/ Malika"]
```

```python
lines = [x for x in linelist]
```

```python
lines = [x for x in lines if x[:5] != 'Album' and x[:6]!='Artist' and x[0]!='*' and x[:5]!='Typed']
```

```python
def replace_verse(x):
    if x[:5]=='Verse' or x[:6]=='Chorus' or x[:5]=='Intro' or x[0]=='[':
        return "VERSE"
    if x[:4]=='Song':
        return "SONG"
    else: 
        return x
```

```python
lines = [replace_verse(x) for x in lines]
```

```python
from nltk.tokenize.sonority_sequencing import SyllableTokenizer
SSP = SyllableTokenizer()
from nltk.tokenize import TreebankWordTokenizer
```

```python
word_lines = [TreebankWordTokenizer().tokenize(x) for x in lines]
```

```python
len(word_lines)
```

```python
syl_lines = [SSP.tokenize(y) for x in word_lines for y in x]
```

```python
vocab = list(set([x for y in word_lines for x in y]))
```

```python
def get_bigrams(sent):
    return [(sent[i],sent[i+1]) for i in range(0,len(sent)-1)]
```

```python
bigrams = [get_bigrams(x) for x in word_lines]
```

```python
bigrams_dict
```

```python
bigrams_dict = dict()
for x in vocab:
    bigrams_dict[x]= []
```

```python
for x in bigrams:
    for y in x:
        bigrams_dict[y[0]].append(y[1])
```

```python
import random

def generate_from_dict(start, dictionary, length):
    res = [start]
    pivot = start
    for x in range(0,length):
        nextw = random.choice(dictionary[pivot])
        pivot = nextw
        res.append(nextw)
    return res
        
    
```

```python
generate_from_dict('We', bigrams_dict, 15)
```
