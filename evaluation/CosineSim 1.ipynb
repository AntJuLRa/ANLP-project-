```python
#split the documents up in words 
holds=[]
#test=songs[:10]
for item in songs:
    holds.append(item)
#print(holds[:20])
allsongs=' '.join(holds)

allwords=allsongs.split()
```

```python
#split each document up in words
wordssongs=[]
for song in songs:
    hold=song.split()
    wordssongs.append(hold)
#print(wordssongs[1])
```

```python
#count all words
allwordcount=Counter(allwords)

#count the words in each document 
songscount=[]
for i in wordssongs:
    songscount.append(Counter(i))
```

```python
#calculates in how many documents each word appears
nj=[]
for key in allwordcount:
    has=0
    for song in songscount:
        if key in song:
            has=has+1
    nj.append(has)
#print(nj)
```

```python
target = open('nj.txt', 'wb')
for item in nj:
    target.write(item.encode('ascii', 'ignore'))
target.close()
```

```python
#the total number of documents
N=len(songs)
print(N)
```

```python
#calculates how often a word appears in each document

freqall=[]

for song in songscount:
    hold=[]
    
    #if word is in the song then lookup how often if not add 0
    for key in allwordcount:
        if key in song:
            val=song.get(key)
            hold.append(val)
        else:
            hold.append(0)
            
    freqall.append(hold) 
            
```

```python
#calculates the inverse document frequency for each word in each document
allwij=[]
for song in freqall:
    songfreq=[]
    for wordfreq,docfreq in zip(song,nj):
        wij=wordfreq*np.log(N/docfreq)
        songfreq.append(wij)
    allwij.append(songfreq)
    
        
    
    
```
